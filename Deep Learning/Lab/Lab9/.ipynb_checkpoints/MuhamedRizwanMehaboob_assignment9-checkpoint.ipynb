## **1. Train a Flow Matching model to generate synthetic data (any one of half moon, swiss roll or circles)**

### a. Implement sinusoidal position encoding for the time input, and a neural network which takes a data point x and time t as inputs to calculate the vector field at that time.
    You should project x using a linear layer to a hidden dimension, and embed t using positional encoding followed by a projection to the hidden dimension. The encoded inputs can then be passed through linear layers with ReLU activation followed by a final projection to the required output dimension.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import sklearn.datasets as ds
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms
```

```python
class SinusoidalTimeEmbedding(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.dim = dim

    def forward(self, t):
        half_dim = self.dim // 2
        emb = torch.exp(torch.arange(half_dim) * (-torch.log(torch.tensor(10000.0)) / (half_dim - 1)))
        emb = t[:, None] * emb[None, :]
        emb = torch.cat([torch.sin(emb), torch.cos(emb)], dim=-1)
        return emb
```

```python
class VectorFieldNet(nn.Module):
    def __init__(self, input_dim=2, hidden_dim=512):
        super().__init__()
        self.x_proj = nn.Linear(input_dim, hidden_dim)
        self.t_embed = SinusoidalTimeEmbedding(hidden_dim)
        self.t_proj = nn.Linear(hidden_dim, hidden_dim)

        self.net = nn.Sequential(
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, input_dim)
        )

    def forward(self, x, t):
        x = self.x_proj(x)
        t = self.t_proj(self.t_embed(t))
        h = x + t
        return self.net(h)
```

### b. Train the Flow Matching model with the above network to generate synthetic data.

```python
def train_flow_model(model, data, epochs=1000, lr=1e-3):
    optimizer = optim.Adam(model.parameters(), lr=lr)
    data = torch.tensor(data, dtype=torch.float32)
    model.train()
    for epoch in range(epochs):
        optimizer.zero_grad()
        t = torch.rand(len(data))
        noise = torch.randn_like(data)
        xt = t[:,None] * data + (1-t)[:, None] * noise
        target = data - noise
        pred = model(xt, t)
        loss = ((pred - target) ** 2).mean()
        loss.backward()
        optimizer.step()
        if epoch % 200 == 0:
            print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
    return model
```

### c. Generate 100 data points by simulating the ODE using Euler method with 1000 steps. Plot the intermediate generated points after every 100 steps to see the noise getting converted to data samples and the final generated data.

```python
def sample_euler(model, steps=1000, num_points=100):
    x = torch.randn(num_points, 2)
    dt = 1.0 / steps
    all_steps = []
    
    for i in range(steps):
        t = torch.ones(num_points) * (i / steps)
        v = model(x, t)
        x = x + v * dt
        if i % 100 == 0:
            all_steps.append(x.detach().cpu().numpy())
    return all_steps, x.detach().cpu().numpy()
```

#### **Plotting for the Make Moons dataset**

```python
data, _ = ds.make_moons(n_samples=1000, noise=0.03)
model = VectorFieldNet()
model = train_flow_model(model, data)

steps_data, final_data = sample_euler(model)
```

```python
for i, step in enumerate(steps_data):
    plt.scatter(step[:, 0], step[:, 1], c = 'red', s=5, label = 'Generated')
    plt.scatter(data[:, 0], data[:, 1], c = 'green', s=5, label = 'Real')
    plt.title(f"Step {i*100}")
    plt.show()
```

```python
plt.scatter(final_data[:, 0], final_data[:, 1], c = 'red', s=5, label = 'Generated')
plt.scatter(data[:, 0], data[:, 1], c = 'green', s=5, label = 'Real')
plt.title("Final Generated Data")
plt.legend()
plt.show()
```

## **2. Train a Flow Matching model to generate MNIST images of any one class.**

### a. Implement a UNet which takes an image and time as inputs to calculate the vector field at that time. The UNet can be 2 levels deep, using SiLU (swish) activation function. At each skip connection, pass a time embedding with the same size as the prepooled feature map in addition to the prepooled feature map to the corresponding expanding path block. (Refer “ Useful Information ” for code snippets to implement the UNet. You have to modify the input sizes and forward pass of the upsampling block and UNET Vector Field to account for the time embedding. Define the model so that the bottleneck layer receives an input with 128 channels and outputs 256 channels.)

```python
class convolution_block(nn.Module):
    def __init__(self, input_channels, output_channels):
        super().__init__()
        #define the operations inside the block
        self.block = nn.Sequential(
            nn.Conv2d(input_channels,output_channels,kernel_size=3,padding=1,bias=False),
            nn.BatchNorm2d(output_channels),
            nn.Conv2d(output_channels,output_channels,kernel_size=3,padding=1,bias=False),
            nn.BatchNorm2d(output_channels),
            nn.SiLU(inplace=True)
        )
    def forward(self, x):
        x = self.block(x)
        return x
```

```python
class downsampling_block(nn.Module):
    def __init__(self, input_channels, output_channels):
        """
        Args:
            input_channels (int): The number of channels in the input feature map.
            output_channels (int): The number of channels after the convolution block.
        """
        super().__init__()
        self.conv_block = convolution_block(input_channels, output_channels)
        self.pool = nn.MaxPool2d((2,2))
    def forward(self,x):
        x = self.conv_block(x)
        p = self.pool(x)
        # Return both the pre-pooled feature map for the skip connection
        # and the pooled feature map for the next layer in the encoder.
        return x,p
```

```python
class upsampling_block(nn.Module):
    """
        An upsampling block for the U-Net expansive path.
        It combines upsampling with a skip connection from the encoder path.
    """
    def __init__(self, input_channels, output_channels):
        super().__init__()
        #define the up-convolution
        self.up_convolution = nn.ConvTranspose2d(input_channels, output_channels,kernel_size=2, stride=2, padding=0)
        #define the convolution: current output + skip connection
        self.convolution_block = convolution_block(output_channels+output_channels,output_channels)
    def forward(self,x,skip_connection):
        x = self.up_convolution(x)
        #concatenate the previous result with the skip connection before applying
        # the convolution block
        x = torch.cat([x,skip_connection],axis=1)
        #apply the convolution
        x = self.convolution_block(x)
        return x
```

```python
class TimeEmbedding(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.embed = SinusoidalTimeEmbedding(dim)
        self.proj = nn.Linear(dim, dim)

    def forward(self, t):
        return self.proj(self.embed(t))
```

```python
class UNET_Vector_Field(nn.Module):
    def __init__(self):
        super().__init__()
        '''
            TIME ENCODINGS FOR THE SKIP CONNECTION
        '''
        self.t1 = TimeEmbedding(128)
        self.t2 = TimeEmbedding(64)
        '''
            CONTRACTING PATH
        '''
        self.encoder_1 = downsampling_block(1,64)
        self.encoder_2 = downsampling_block(64,128)
        '''
            BOTTLENECK
        '''
        self.bottleneck = convolution_block(128,256)
        '''
            EXPANSIVE PATH
        '''
        self.decoder_1 = upsampling_block(256,128)
        self.decoder_2 = upsampling_block(128,64)
        '''
            OUTPUT LAYER
        '''
        self.output_layer = nn.Conv2d(64,1,kernel_size=1)
    
    def forward(self,x,t):
        #contracting path
        x1,p1 = self.encoder_1(x)
        x2,p2 = self.encoder_2(p1)
        #bottleneck
        b = self.bottleneck(p2)
        #expanding path
        # Add time encodings here

        t1 = self.t1(t).unsqueeze(-1).unsqueeze(-1)
        t2 = self.t2(t).unsqueeze(-1).unsqueeze(-1)

        x2 = x2 + t1
        x1 = x1 + t2
        
        d1 = self.decoder_1(b,x2) #use of skip connection
        d2 = self.decoder_2(d1,x1) #use of skip connection
        #output layer
        output = self.output_layer(d2)
        return output
```

### b. Train the Flow Matching model with the above network to generate images. Train for at least 3 epochs. 

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

mnist = datasets.MNIST(root="", train=True, download=False, transform=transform)
```

```python
target_class = 0
indices = [i for i, t in enumerate(mnist.targets) if t== target_class]
subset = Subset(mnist, indices)
train_dataloader = DataLoader(subset, batch_size=64, shuffle=True)
```

```python
model_img = UNET_Vector_Field()
optimizer = optim.Adam(model_img.parameters(),lr=1e-3)
```

```python
for epoch in range(3):
    for x, _ in train_dataloader:
        t = torch.rand(x.size(0))
        noise = torch.randn_like(x)
        xt = t[:, None, None, None] * x + (1 - t)[:, None, None, None] * noise
        target = x - noise
        pred = model_img(xt, t)
        loss = ((pred - target) ** 2).mean()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1} done, Loss: {loss.item():.4f}")
```

### c. Generate an image from noise by simulating the ODE using Euler method with 1000 steps. Display the intermediate image after every 100 steps during generation for a single image to see the noise getting converted into a data sample.

```python
def sample(model):
    x = torch.randn(1,1,28,28)
    for i in range(1000):
        t = torch.ones(1) * (i/1000)
        x = x + model(x, t) * (1/1000)

        if i % 100 == 0:
            plt.imshow(x[0,0].detach().cpu(), cmap='gray')
            plt.title(f"Step: {i}")
            plt.show()

sample(model_img)
```
